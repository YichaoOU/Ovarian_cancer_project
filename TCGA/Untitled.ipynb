```python
def df_NCV(df):

	Y = df.Class
# 	print (Y)
	y_true = []
	y_pred = []
	X = df.drop('Class',axis=1)
	skf_outer = StratifiedKFold(Y, n_folds=5,shuffle=True)
# 	print "skf_outer",skf_outer
	tuned_parameters_dt = {'min_samples_split': [2,5],'max_depth':[2,5,30]}
	acc_array=[]
	dv_tree_count = 1
	for train_index_outer, test_index_outer in skf_outer:
		train_y_outer = Y[train_index_outer]
		size = len(train_y_outer)
		print (size)
		train_x_outer = X.iloc[train_index_outer]
		test_y_outer = Y[test_index_outer]
		test_x_outer = X.iloc[test_index_outer]
# 		print "train_y_outer",train_y_outer
# 		print "test_y_outer",test_y_outer
# 		dt_clf = GridSearchCV(DTC(criterion='gini'), tuned_parameters_dt,scoring="accuracy",cv=5)
		dt_clf = GridSearchCV(RFC(criterion='entropy',n_estimators=40,class_weight={'S2':10000,"S3":1,"S4":10000}), tuned_parameters_dt,scoring="accuracy",cv=5)
		dt_clf.fit(train_x_outer, train_y_outer)

# 		print "inner CV result"
		count = 0
		for item in dt_clf.grid_scores_:
			count += 1
# 			print count,item
# 		print "inner CV best"
		print (dt_clf.best_score_) 
		y_true_dt, y_pred_dt = test_y_outer, dt_clf.predict(test_x_outer)
		y_true+=y_true_dt.tolist()
		y_pred+=list(y_pred_dt)		
# 		new_dt = DTC()
# 		dot_file_name = "LMP_PSO_DT_"+str(dv_tree_count)+".dot"
# 		f=open(dot_file_name,"wb")
# 		export_graphviz(dt_clf.best_estimator_ , out_file=dot_file_name,feature_names=X.columns.tolist())
# 		dv_tree_count += 1
# 		tn, fp, fn, tp = confusion_matrix(y_true_dt, y_pred_dt).ravel()
		accuracy = accuracy_score(y_true_dt, y_pred_dt)
		acc_array.append(accuracy)
	return acc_array,mean(acc_array),y_true,y_pred
a,b,c,d = df_NCV(LMP_PSO_df[LMP_PSO_df['Class'].isin(['S3','S2','S4'])])
```
