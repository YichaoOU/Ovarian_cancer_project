```python
df = pd.read_csv("union_sdegs_expression_file.csv",index_col=0)
# df = df.drop(['Unnamed: 13'],axis=1)
df.head()
```

```python
import scipy
def get_columns(cols,s):
    out = []
    for c in cols:
        if s in c:
            out.append(c)
    return out
def p_value(r,a,b):
    a_list = [r[i] for i in a]
    b_list = [r[i] for i in b]
    return stats.ttest_ind(a_list,b_list).pvalue
LMP_cols = get_columns(df.columns,"LMP")
PSO2_cols = get_columns(df.columns,"PSO2")
PSO_cols = get_columns(df.columns,"PSO")
PSO3_cols = get_columns(df.columns,"PSO3")
df['LMP_PSO'] = df.apply(lambda r:p_value(r,LMP_cols,PSO_cols),axis=1)
df['PSO2_PSO3'] = df.apply(lambda r:p_value(r,PSO2_cols,PSO3_cols),axis=1)
```
